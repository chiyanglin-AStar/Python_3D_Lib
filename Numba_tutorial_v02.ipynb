## Numba 介紹

### Subtask:
簡要介紹 Numba 是什麼、它的工作原理（即時編譯 JIT）以及它為何能加速 Python 數值運算。


Numba 是一個開源的 JIT (Just-In-Time) 編譯器，它可以將 Python 函數編譯成機器碼，從而顯著提高數值運算的執行速度。它主要針對數值計算的 Python 程式碼進行優化。

### Numba 的工作原理：即時編譯 (JIT)
當你使用 Numba 裝飾器 (`@jit`) 裝飾一個 Python 函數時，Numba 會在該函數**第一次被呼叫時**將其編譯。這意味著：
1.  **編譯階段**：Numba 會分析你的 Python 程式碼，並將其轉換成優化後的機器碼。這個過程只發生一次。
2.  **執行階段**：之後每次呼叫同一個函數時，Numba 都會直接執行已經編譯好的機器碼，而不是重新解釋 Python 程式碼。

### Numba 為何能加速 Python 數值運算？
Numba 之所以能大幅加速 Python 數值運算，主要有以下幾個原因：
*   **繞過 Python 解釋器開銷**：Python 是一種解釋型語言，每次執行程式碼時都需要解釋器進行解析和執行。Numba 將 Python 函數直接編譯成機器碼，消除了每次執行時的解釋器開銷，使得程式碼能夠直接在 CPU 上運行。
*   **利用底層優化**：Numba 可以利用 LLVM (Low Level Virtual Machine) 編譯基礎設施，對生成的機器碼進行底層優化，例如：
    *   **循環優化**：Numba 對於數值計算中常見的 `for` 循環有非常好的優化效果。
    *   **SIMD (Single Instruction, Multiple Data) 指令**：現代 CPU 支援 SIMD 指令，允許一次操作多個數據。Numba 可以自動或半自動地利用這些指令來加速數組操作。
    *   **記憶體訪問優化**：Numba 可以更好地管理記憶體訪問模式，提高數據存取的效率。
*   **與 NumPy 高度整合**：Numba 與 NumPy 庫緊密結合，可以高效地編譯包含 NumPy 函數的程式碼，並將 NumPy 陣列操作轉換為快速的機器碼。

## 基本 JIT 加速示範

在本節中，我們將示範 Numba 如何透過即時編譯 (JIT) 來加速 Python 數值運算函數。我們將會定義一個簡單的數學函數，並比較其原始版本與經由 Numba `@jit` 裝飾器加速後的版本。

**Reasoning**:
The first instruction is to define a simple Python function that calculates the sum of squares without using Numba. I will use a code block for this.



````python
import numpy as np

def sum_of_squares_python(numbers):
    """Calculates the sum of squares of numbers in a list using pure Python."""
    total = 0
    for num in numbers:
        total += num * num
    return total

print("Python function 'sum_of_squares_python' defined.")
````

**Reasoning**:
The next instruction is to create a JIT-accelerated version of the sum of squares function using the `numba.jit` decorator. I will define this function in a new code block.



````python
from numba import jit

@jit
def sum_of_squares_numba(numbers):
    """Calculates the sum of squares of numbers using Numba JIT acceleration."""
    total = 0
    for num in numbers:
        total += num * num
    return total

print("Numba JIT function 'sum_of_squares_numba' defined.")
````

## 效能比較：有 Numba vs. 無 Numba

### Subtask:

**Reasoning**:
Now I will import the `time` module as instructed to enable precise measurement of function execution times, and then generate a large list of numbers using NumPy for performance testing, which is the first and second part of the instructions.



````python
import time

# Generate a large list of numbers for performance testing
large_numbers = np.random.randint(0, 100, size=10**7) # 10 million random integers

print(f"Generated a list of {len(large_numbers)} numbers for testing.")
````

**Reasoning**:
I will now measure the execution time of both the pure Python function and the Numba-accelerated function using the `time` module, as per instructions 3 and 4 of the subtask. I will also pre-warm the Numba function to ensure JIT compilation occurs before timing the main execution.



````python
print("\n--- Performance Comparison ---")

# --- Measure pure Python function performance ---
start_time = time.time()
result_python_large = sum_of_squares_python(large_numbers)
end_time = time.time()
time_python = end_time - start_time
print(f"Pure Python function execution time: {time_python:.4f} seconds")

# --- Measure Numba JIT-accelerated function performance ---
# Numba needs to compile on the first call. Call once with small data to pre-warm.
_ = sum_of_squares_numba(np.array([1, 2, 3]))

start_time = time.time()
result_numba_large = sum_of_squares_numba(large_numbers)
end_time = time.time()
time_numba = end_time - start_time
print(f"Numba JIT function execution time (after warm-up): {time_numba:.4f} seconds")

# Verify results for large data (optional, but good for sanity check)
# Note: large_numbers is a numpy array, sum_of_squares_python expects a list.
# Let's convert large_numbers to a list for sum_of_squares_python to ensure fair comparison if originally designed for lists.
# However, sum_of_squares_python can iterate over numpy arrays directly, so no conversion needed for functionality.
# For Numba, it's often more efficient with numpy arrays, which is what large_numbers is.

# Let's ensure the results are the same
if result_python_large == result_numba_large:
    print("Results for large datasets match.")
else:
    print("Warning: Results for large datasets do NOT match.")

print(f"Acceleration factor: {time_python / time_numba:.2f}x")
````

## 總結 Numba `nopython=True` 的重要性

`nopython=True` 是 Numba `@jit` 裝飾器的一個關鍵參數，它強制 Numba 以「nopython」模式編譯函數。當設置 `nopython=True` 時，Numba 會嘗試將整個 Python 函數編譯成機器碼，而不允許任何 Python 解釋器回退（fallback）。這對於最大化 Numba 的效能和啟用早期錯誤檢測至關重要。

### 1. 最大化 Numba 的加速效能

*   **徹底擺脫 Python 解釋器**：在 `nopython=True` 模式下，Numba 不會允許函數內部有任何操作回退到 Python 解釋器。這意味著整個函數都將以高效的機器碼執行，徹底消除了 Python 物件模型和解釋器開銷。
*   **類型推斷和專業化**：Numba 在 nopython 模式下能夠進行更積極的類型推斷和函數專業化。它會為輸入參數和中間變量推斷出精確的機器類型（例如 `int32`, `float64`），並根據這些類型生成高度優化的機器碼。如果無法推斷出類型或遇到 Numba 不支援的 Python 特性，編譯會失敗，而不是默默地回退到低效率的物件模式。
*   **啟用更深層次的優化**：由於 Numba 確切知道所有數據類型和操作，LLVM 後端可以執行更激進的優化，例如自動向量化 (SIMD)、循環展開 (loop unrolling) 和記憶體訪問優化，這些在混合模式下通常無法實現。

### 2. 啟用早期錯誤檢測

*   **編譯時錯誤而非運行時錯誤**：當 `nopython=True` 時，如果 Numba 無法將函數的某個部分編譯為機器碼（例如，使用了 Numba 不支援的 Python 庫函數，或者進行了不兼容的類型操作），它會在**編譯階段**立即拋出 `TypingError` 或 `NumbaValueError` 等錯誤。這比在運行時才發現問題要好得多。
*   **強制寫出 Numba 友好的程式碼**：由於 `nopython=True` 的嚴格性，它鼓勵（或強制）開發者編寫更「Numba 友好」的程式碼，即主要使用數值運算和 NumPy 陣列操作的程式碼。這有助於避免不經意間引入導致效能低下的 Python 物件操作。
*   **透明性和可預測性**：通過強制完全編譯，`nopython=True` 提供了更透明和可預測的效能。你明確知道你的程式碼是否完全被 Numba 加速了，而不會有部分程式碼默默地回退到慢速的 Python 解釋器。

總之，`nopython=True` 是使用 Numba 實現最大效能和確保程式碼健康的重要工具。它將 Numba 從一個潛在的加速器變成一個可靠的高效能計算引擎。

### Numba 如何與 NumPy 高效整合？

1.  **自動理解 NumPy 陣列**：當 Numba 編譯一個函數時，它能夠自動識別輸入和輸出的 NumPy `ndarray` 物件及其數據類型（`dtype`）和形狀（`shape`）。Numba 會將這些 Python 物件轉換為底層的、類型化的機器碼表示，這使得直接在記憶體中的數據上執行操作成為可能。

3.  **直接呼叫 NumPy 內部函數**：Numba 能夠直接使用 NumPy 內部高度最佳化且支援的函數（如 `np.sum`, `np.dot`, `np.sqrt` 等），並將它們編譯為底層指令。這使得在 Numba JIT 編譯的函數中，NumPy 函數的呼叫成本極低。

The next step is to define a Numba-accelerated version of the dot product function using `@njit` (equivalent to `@jit(nopython=True)`), as it is crucial for maximizing Numba's performance with NumPy arrays.



````python
from numba import njit
import numpy as np

@njit
def dot_product_numba(arr1, arr2):
    """Calculates the dot product of two NumPy arrays using Numba's @njit acceleration."""
    # In nopython mode, Numba directly optimizes NumPy operations.
    # np.dot is typically well-optimized by Numba for supported types.
    return np.dot(arr1, arr2)

print("Numba JIT (nopython=True) function 'dot_product_numba' defined.")
````

### Numba's Core Advantages and `nopython=True`

Numba acts as a Just-In-Time (JIT) compiler, transforming Python functions into optimized machine code. This process bypasses the Python interpreter's overhead, allowing for execution speeds comparable to compiled languages like C or Fortran.

The `@jit` decorator is fundamental, but the more strict `@njit` (equivalent to `@jit(nopython=True)`) is crucial for maximizing performance. The `nopython=True` mode forces Numba to compile the entire function into machine code without allowing any fallback to the Python interpreter. This enables aggressive type inference, deep LLVM-based optimizations (like automatic vectorization and loop unrolling), and provides early error detection at compile-time (`TypingError`, `NumbaValueError`) rather than runtime. This encourages developers to write "Numba-friendly" code, focusing on numerical operations and array manipulations.

However, `nopython=True` comes with limitations:
*   **Dynamic Typing Restrictions**: It requires static type inference, which can conflict with Python's dynamic nature.
*   **Limited Python Feature Support**: Complex Python features like custom classes/objects, intricate recursion, certain exception handling scenarios, and standard Python iterators/generators may not be fully supported.
*   **External Library Challenges**: Most pure Python standard libraries (e.g., `datetime`, `re`) and non-numerical third-party libraries are not directly callable within `nopython` mode, necessitating code restructuring.

Numba's strength is particularly evident in its deep integration with NumPy.
````python
from numba import cuda

@cuda.jit
def gpu_add(x, y, out):
    """Perform element-wise addition of two arrays on the GPU."""
    idx = cuda.grid(1)
    if idx < x.size:
        out[idx] = x[idx] + y[idx]

print("GPU kernel function 'gpu_add' defined.")
````

## CUDA 執行階層：Grid, Block, Thread 模型

NVIDIA GPU 是一個高度平行的處理器，其架構的設計旨在同時處理數千個甚至數百萬個計算任務。為了有效地在 GPU 上執行程式碼（稱為 **Kernel 函數**），CUDA 引入了一個三層次的執行階層模型：**Thread (執行緒)**、**Block (區塊)** 和 **Grid (網格)**。

理解這個模型對於編寫高效能的 GPU 程式碼至關重要。

### 1. Thread (執行緒)

*   **代表**：Thread 是 CUDA 程式設計中最小的執行單元。每個 Thread 都會執行 Kernel 函數中的程式碼。
*   **作用**：在 Kernel 內部，每個 Thread 都有其獨立的程式計數器、暫存器和堆疊，但它們通常會共享指令碼。Thread 的主要任務是處理數據的一個獨立部分，例如陣列中的一個元素。
*   **識別**：每個 Thread 在其所屬的 Block 內都有一個唯一的索引（`threadIdx.x`, `threadIdx.y`, `threadIdx.z`）。
*   **平行性**：Thread 是高度平行的，數百萬個 Thread 可以同時運行，每個處理自己的數據片段。然而，Thread 的執行是輕量級的，切換成本低。

### 2. Block (執行緒區塊)

*   **代表**：Block 是一組具有相同程式碼、由 CUDA 硬體調度並行執行的 Thread 的集合。所有在同一個 Block 裡的 Thread 可以透過共享記憶體和同步原語（例如 `__syncthreads()`）互相協調合作。
*   **作用**：Block 內的 Thread 之間可以協同工作、共享數據，這對於實現更複雜的平行演算法非常有用。它們被設計為一起執行，並且可以對共享數據進行快速存取。
*   **識別**：每個 Block 在其所屬的 Grid 內都有一個唯一的索引（`blockIdx.x`, `blockIdx.y`, `blockIdx.z`）。
*   **限制**：一個 Block 內部的 Thread 數量是有限制的（通常最多 1024 個 Thread），且所有 Thread 必須在同一串流處理器 (SM) 上執行，這意味著它們共享 SM 的資源。

### 3. Grid (執行緒網格)

*   **代表**：Grid 是由一個或多個 Block 組成的集合。一個 Kernel 函數的所有計算任務都是在一個 Grid 中完成的。
*   **作用**：Grid 抽象了整個計算任務，它將數據集分割成獨立的 Block，每個 Block 可以獨立地調度到 GPU 上的任何可用的串流處理器 (SM) 上執行。不同 Block 之間的執行是完全獨立的，沒有隱含的執行順序或數據共享機制（除非透過全局記憶體）。
*   **識別**：Grid 沒有自己的索引，但可以通過 `gridDim.x`, `gridDim.y`, `gridDim.z` 來表示 Grid 的維度（即包含多少個 Block）。
*   **平行性**：Grid 是最高層次的平行抽象，它允許 Kernel 擴展到任何大小的 GPU，無論 GPU 有多少個串流處理器。只要 GPU 有足夠的資源，它就可以將 Grid 中的 Block 調度到不同的 SM 上並行執行。

### 階層關係與協同工作

這三層形成了一個強大的階層結構，允許開發者將複雜的平行問題映射到 GPU 硬體上：

```text
Grid (整個計算任務)
  └── Block (一組協同工作的 Threads)
        └── Thread (最小的執行單元，執行 Kernel 程式碼)
```

當一個 Kernel 函數被啟動時，程式會指定 Grid 和 Block 的維度（例如 `kernel[grid_dim, block_dim](args)`）。

*   每個 **Thread** 計算自己的全局索引（例如使用 `cuda.grid(1)`），然後處理輸入數據中與該索引對應的元素。這就是數據如何在 GPU 上被「切片」並由不同 Thread 處理的方式。
*   多個 **Thread** 被組織成一個 **Block**。Block 內部的 Thread 可以透過共享記憶體和同步點來交換中間結果，這對許多演算法（如歸約、前綴和）至關重要。
*   多個 **Block** 被組織成一個 **Grid**。Grid 允許我們處理任意大小的數據集，而無需關心 GPU 內部有多少個物理處理單元。CUDA 運行時會自動將 Grid 中的 Block 分配給 GPU 的 SM 執行。

這種模型使得開發者可以專注於編寫平行演算法的邏輯，而將底層的硬體調度和資源管理交由 CUDA 運行時系統處理。

*   **平行性**：Thread 是高度平行的，數百萬個 Thread 可以同時運行，每個處理自己的數據片段。然而，Thread 的執行是輕量級的，切換成本低。

### 2. Block (執行緒區塊)

*   **代表**：Block 是一組具有相同程式碼、由 CUDA 硬體調度並行執行的 Thread 的集合。所有在同一個 Block 裡的 Thread 可以透過共享記憶體和同步原語（例如 `__syncthreads()`）互相協調合作。
*   **作用**：Block 內的 Thread 之間可以協同工作、共享數據，這對於實現更複雜的平行演算法非常有用。它們被設計為一起執行，並且可以對共享數據進行快速存取。
*   **識別**：每個 Block 在其所屬的 Grid 內都有一個唯一的索引（`blockIdx.x`, `blockIdx.y`, `blockIdx.z`）。
*   **限制**：一個 Block 內部的 Thread 數量是有限制的（通常最多 1024 個 Thread），且所有 Thread 必須在同一串流處理器 (SM) 上執行，這意味著它們共享 SM 的資源。

### 3. Grid (執行緒網格)

*   **代表**：Grid 是由一個或多個 Block 組成的集合。一個 Kernel 函數的所有計算任務都是在一個 Grid 中完成的。
*   **作用**：Grid 抽象了整個計算任務，它將數據集分割成獨立的 Block，每個 Block 可以獨立地調度到 GPU 上的任何可用的串流處理器 (SM) 上執行。不同 Block 之間的執行是完全獨立的，沒有隱含的執行順序或數據共享機制（除非透過全局記憶體）。

### 階層關係與協同工作

這三層形成了一個強大的階層結構，允許開發者將複雜的平行問題映射到 GPU 硬體上：

## CUDA 執行階層：Grid, Block, Thread 模型

NVIDIA GPU 是一個高度平行的處理器，其架構的設計旨在同時處理數千個甚至數百萬個計算任務。為了有效地在 GPU 上執行程式碼（稱為 **Kernel 函數**），CUDA 引入了一個三層次的執行階層模型：**Thread (執行緒)**、**Block (區塊)** 和 **Grid (網格)**。

理解這個模型對於編寫高效能的 GPU 程式碼至關重要。

### 1. Thread (執行緒)

*   **代表**：Thread 是 CUDA 程式設計中最小的執行單元。每個 Thread 都會執行 Kernel 函數中的程式碼。
*   **作用**：在 Kernel 內部，每個 Thread 都有其獨立的程式計數器、暫存器和堆疊，但它們通常會共享指令碼。Thread 的主要任務是處理數據的一個獨立部分，例如陣列中的一個元素。
*   **識別**：每個 Thread 在其所屬的 Block 內都有一個唯一的索引（`threadIdx.x`, `threadIdx.y`, `threadIdx.z`）。
*   **平行性**：Thread 是高度平行的，數百萬個 Thread 可以同時運行，每個處理自己的數據片段。然而，Thread 的執行是輕量級的，切換成本低。

### 2. Block (執行緒區塊)

*   **代表**：Block 是一組具有相同程式碼、由 CUDA 硬體調度並行執行的 Thread 的集合。所有在同一個 Block 裡的 Thread 可以透過共享記憶體和同步原語（例如 `__syncthreads()`）互相協調合作。
*   **作用**：Block 內的 Thread 之間可以協同工作、共享數據，這對於實現更複雜的平行演算法非常有用。它們被設計為一起執行，並且可以對共享數據進行快速存取。
*   **識別**：每個 Block 在其所屬的 Grid 內都有一個唯一的索引（`blockIdx.x`, `blockIdx.y`, `blockIdx.z`）。
*   **限制**：一個 Block 內部的 Thread 數量是有限制的（通常最多 1024 個 Thread），且所有 Thread 必須在同一串流處理器 (SM) 上執行，這意味著它們共享 SM 的資源。

### 3. Grid (執行緒網格)

*   **代表**：Grid 是由一個或多個 Block 組成的集合。一個 Kernel 函數的所有計算任務都是在一個 Grid 中完成的。
*   **作用**：Grid 抽象了整個計算任務，它將數據集分割成獨立的 Block，每個 Block 可以獨立地調度到 GPU 上的任何可用的串流處理器 (SM) 上執行。不同 Block 之間的執行是完全獨立的，沒有隱含的執行順序或數據共享機制（除非透過全局記憶體）。
*   **識別**：Grid 沒有自己的索引，但可以通過 `gridDim.x`, `gridDim.y`, `gridDim.z` 來表示 Grid 的維度（即包含多少個 Block）。

*   **作用**：在 Kernel 內部，每個 Thread 都有其獨立的程式計數器、暫存器和堆疊，但它們通常會共享指令碼。Thread 的主要任務是處理數據的一個獨立部分，例如陣列中的一個元素。
*   **識別**：每個 Thread 在其所屬的 Block 內都有一個唯一的索引（`threadIdx.x`, `threadIdx.y`, `threadIdx.z`）。
*   **平行性**：Thread 是高度平行的，數百萬個 Thread 可以同時運行，每個處理自己的數據片段。然而，Thread 的執行是輕量級的，切換成本低。

### 2. Block (執行緒區塊)

*   **代表**：Block 是一組具有相同程式碼、由 CUDA 硬體調度並行執行的 Thread 的集合。所有在同一個 Block 裡的 Thread 可以透過共享記憶體和同步原語（例如 `__syncthreads()`）互相協調合作。
*   **作用**：Block 內的 Thread 之間可以協同工作、共享數據，這對於實現更複雜的平行演算法非常有用。它們被設計為一起執行，並且可以對共享數據進行快速存取。
*   **識別**：每個 Block 在其所屬的 Grid 內都有一個唯一的索引（`blockIdx.x`, `blockIdx.y`, `blockIdx.z`）。
*   **限制**：一個 Block 內部的 Thread 數量是有限制的（通常最多 1024 個 Thread），且所有 Thread 必須在同一串流處理器 (SM) 上執行，這意味著它們共享 SM 的資源。

### 3. Grid (執行緒網格)

*   **代表**：Grid 是由一個或多個 Block 組成的集合。一個 Kernel 函數的所有計算任務都是在一個 Grid 中完成的。
*   **作用**：Grid 抽象了整個計算任務，它將數據集分割成獨立的 Block，每個 Block 可以獨立地調度到 GPU 上的任何可用的串流處理器 (SM) 上執行。不同 Block 之間的執行是完全獨立的，沒有隱含的執行順序或數據共享機制（除非透過全局記憶體）。
*   **識別**：Grid 沒有自己的索引，但可以通過 `gridDim.x`, `gridDim.y`, `gridDim.z` 來表示 Grid 的維度（即包含多少個 Block）。
*   **平行性**：Grid 是最高層次的平行抽象，它允許 Kernel 擴展到任何大小的 GPU，無論 GPU 有多少個串流處理器。只要 GPU 有足夠的資源，它就可以將 Grid 中的 Block 調度到不同的 SM 上並行執行。

### 階層關係與協同工作

這三層形成了一個強大的階層結構，允許開發者將複雜的平行問題映射到 GPU 硬體上：

```text
Grid (整個計算任務)
  └── Block (一組協同工作的 Threads)
        └── Thread (最小的執行單元，執行 Kernel 程式碼)
```

當一個 Kernel 函數被啟動時，程式會指定 Grid 和 Block 的維度（例如 `kernel[grid_dim, block_dim](args)`）。

*   每個 **Thread** 計算自己的全局索引（例如使用 `cuda.grid(1)`），然後處理輸入數據中與該索引對應的元素。這就是數據如何在 GPU 上被「切片」並由不同 Thread 處理的方式。
*   多個 **Thread** 被組織成一個 **Block**。Block 內部的 Thread 可以透過共享記憶體和同步點來交換中間結果，這對許多演算法（如歸約、前綴和）至關重要。
*   多個 **Block** 被組織成一個 **Grid**。Grid 允許我們處理任意大小的數據集，而無需關心 GPU 內部有多少個物理處理單元。CUDA 運行時會自動將 Grid 中的 Block 分配給 GPU 的 SM 執行。

這種模型使得開發者可以專注於編寫平行演算法的邏輯，而將底層的硬體調度和資源管理交由 CUDA 運行時系統處理。


## CUDA 執行階層：Grid, Block, Thread 模型

NVIDIA GPU 是一個高度平行的處理器，其架構的設計旨在同時處理數千個甚至數百萬個計算任務。為了有效地在 GPU 上執行程式碼（稱為 **Kernel 函數**），CUDA 引入了一個三層次的執行階層模型：**Thread (執行緒)**、**Block (區塊)** 和 **Grid (網格)**。

Following this, all cells from the beginning of the 'Numba CUDA' section should be re-executed to ensure proper environment setup and re-check GPU availability.
